```python
import requests
import json
import sys
 
service_name = "js-eus-search"
key = ""
service_endpoint = "https://js-eus-search.search.windows.net"
index = "vector-1711569798020"
 
def update_url(chunk_id, url):
 
    base_url=f"https://{service_name}.search.windows.net/indexes/{index}/docs/index?api-version=2020-06-30"
 
    headers = {"Content-Type": "application/json", "api-key": key}
 
    body = {
    "value": [
        {
            "@search.action": "merge",
            "chunk_id": chunk_id,
            "url": url
        }
    ]
    }
    results = requests.post(url = base_url, headers = headers, json = body)
```

```python
from azure.core.credentials import AzureKeyCredential
from azure.search.documents import SearchClient
 
 
search_client = SearchClient(service_endpoint, index, AzureKeyCredential(key))
```

```python
results = search_client.search(search_text="*")
 
for result in results:
    chunk_id = result['chunk_id']
    title = result['title']
    url = 'https://alberta.ca/{}'.format(title[:-5])
    update_url(chunk_id, url)
```
